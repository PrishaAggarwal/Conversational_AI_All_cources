```python
#sabse phele information gain we have to find since voh root node banega
def find_entropy(df):
  target=df.keys()[-1]
  target_values = df[target].unique()
  entropy = 0
  for value in target_values:
    prob=(df[df[target]==value])/len(df)
    entropy += -prob*np.log2(prob)
  return entropy
```

```python
find_entropy(df)
```

```python
def avg_info_entropy(df,attribute):
  attribute_values = df[attribute].unique()
  avg_info_entopy = 0
  for value in attribute_values:
    sub_df = df[df[attribute]==value]
    entropy = find_entropy(sub_df)
    weight=len(sub_df)/len(df)
    avg_info_entropy += weight*entropy
  return avg_info_entropy
```

```python
avg_info_entropy(df,'Outlook')
```

```python
#yeh sab functions humne 9th mein bhi banaye hai...aab gain ratio krna hai
#har ek attribute ka humne nikalna hai to find attribute for each level using information gain

def find_winner(df):
  attributes=df.keys()[:-1]
  IG=[]
  for attribute in attributes:
    IG.append(find_entropy(df)-avg_info_entropy(df,attribute))
  return attributes[np.argmax[IG]]
```

```python
find_winner(df) #har level par yeh function call hoga and ek attribute choose ho jayega with max information gain
#argmax is index value of that attribute jo upar wala function is returning
```

```python
def build_tree(df):
  #sabse phele agar sab class samples ek hi class ko belong krte hai then udhr vhi class jayega os attribute ke niche
  target = df.keys()[-1]
  #now agar unique value is just 1 then koi bhi row ki value return krvado kyuki sab value same hai
  #ya phir yeh unique value save krke return krdo
  if len(df[target].unique())==1:
    return df[target].iloc[0]
  #agar target variable hai bas column and no other column is left
  #yeh especially tab hota hai jab subsets bana rhe ho and we are left with bas target variable
  #yeh random forest mein bahut hota hai
  #toh then target values mein jo jiyada hoga suppose yes then vhi return ho jayega
  if len(df.columns)==1:
    return df[target].mode()[0]
  #yeh above 2 base case hote hai and we should remember these
  best_feature = find_winner(df)
  tree={best_feature:{}}
  #making use of dictionary to make the tree
  #aab jo best feature hai oski unique values chahiye
  best_feature_values = df[best_feature].unique()
  #aab ek ek krke har attribute lenge and oske under subsample mein subdf nikalenge
  for value in best_feature_values:
    sub_df = df[df[best_feature]==value]
    if sub_df.empty:
      #i.e bas target values left and no columns
      #then again target wale ka mode de denge
      tree[best_feature][value]=df[target].mode()[0]
    else:
      tree[best_feature][value]=build_tree(sub_df) #again os subsample par yeh function call kr rhe hai
  return tree
```

```python
#ek library hia jo dictionary ko parse krke print krva dega i.e pprint
tree=build_tree(df)
import pprint
pprint.pprint(tree)
```

```python
#now doing prediction from this tree
def prediction(tree,test_instance):
  #test_instance voh exmaple hai jiska label nikalna hai
  #we have to traverse the tree
  #sabse phele check tree and not test_instance
  #there is a chance that voh test_instance mein jo attibute hai voh not in tree jaise temp
  for attribute in tree.keys():
    #for example if u see in dictionary then outlook is key for whole tree
    value=test_instance[attribute]
    #check in tree that os attribute ke under kya value hai
    sub_tree=tree[attribute][value]
    #sub_tree can be an attribute or dictionary i.e oske aage aur cheeze hai
    if type(sub_tree) is dict:
      label=prediction(sub_tree,test_instance)
    else: #i.e voh ek value hai then sub_tree is oska label hai
      label=sub_tree
      break;
    return label
```

```python
#For random forest just take subset  of features and bootstrap sample and make alot of tree
def bootstrap(df):
  #random rows jismein rows can be repeated
  #df ki jo total length hai osmein 80 percent leli
  #20 percent we put out of bag
  n_samples=int(len(df)*0.8)
  #this 80 percent is called bootstrap ratio
  return df.sample(n_samples,replace=True)
```

```python
#ek aur feature aab for subset of features
#find winner modify since subset of feature hia hamare pass
#we have sed subset_features=None means it is optional
def find_winner(df,subset_features=None):
  attributes=df.keys()[:-1]
  if subset_features is not None:
    #phele hum sab features ka IG nikal rhe the but aab hum bas vhi attributes pick krenge jo subset features mein hai
    attributes=[att for att in attributes in ]
  IG=[]
  for attribute in attributes:
    IG.append(find_entropy(df)-avg_info_entropy(df,attribute))
  return attributes[np.argmax[IG]]
```

```python
#similarly hum ismein bhi change krnge i.e build_tree
def build_random_forest(df,feature_subset):
  target = df.keys()[-1]
  if len(df[target].unique())==1:
    return df[target].iloc[0]
  if len(df.columns)==1 or (feature_subset is not None and len(feature_subset)==0):
    return df[target].mode()[0]
  best_feature = find_winner(df,feature_subset)
  tree={best_feature:{}}
  best_feature_values = df[best_feature].unique()
  for value in best_feature_values:
    sub_df = df[df[best_feature]==value]
    if sub_df.empty:
      tree[best_feature][value]=df[target].mode()[0]
    else:
      new_features=None
      if feature_subset is not None:
        new_features=[f for f in feature_subset if f != best_feature]
      tree[best_feature][value]=build_random_forest(sub_df)
  return tree
```

```python
import random
def random_forest(df,n_trees,max_features=None):
  trees=[]
  features=df.keys()[-1]
  for i in range(n_trees):
    #har tree ke liye diff bootstrapping hogi
    #aab kitne features hone chahiye
    #iska ek rule bhi hia i.e sqroot of total features
    sample_df=bootstrap(df)
    if max_features is None:
      max_features=int(np.sqrt(len(features))) #aab jo features humne nikalne hai onko randomly nikalna hai using random library
    subset_features=random.sample(features,max_features)
    trees.append(build_random_forest(sample_df,subset_features))
  return trees
```
