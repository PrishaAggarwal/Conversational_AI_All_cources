```python
#web scraping
#1. we request the website to give rsw index.html file
#2. we get it
#3. web broswer conberys the html to webpage
#we use web scraping libraries in between the step 2 and step 3
#a tree is created by beautiful soup library ...the tree is of structure of website(html)
#it also provides to traverse throughout the tree

#!pip install request
#!pip install bs4 #bs4 is latest beautiful soup package

import requests
from bs4 import BeautifulSoup
```

```python
#suppose we will scratch data from website..amazon bestsellers books
#books are numbered and each page has around 50 books
#each cell is a location on website
#whereever there is book information,it is some kind of cell
#suppose we want to scrap the bestselling books in a location ...we need to make database with price name of book etc.
#first we want to get the url
#if u want 1st page then  upar wala url is ohk but multiple pages ke  liye aise work nhi krega
#we saw page 2 ka url kya hai and we are going to make a general function to get data from page number

def get_data(pageno):
  url="https://www.amazon.in/gp/bestsellers/books/ref=zg_bs_pg_"+str(pageno)+'_books?ie=UTF8&pg='+str(pageno) #pageno we pass in form of string...humne ending of link hta diya tha
  html=requests.get(url) #requests is used to html code from website
  #now lets create soup of this code...soup is tree like structure of code
  #html.content gives raw content of website
  soup=BeautifulSoup(html.content,'html.parser') #parser is used to create a tree like stucture...tree banane ke we always need a algo...all rules to make tree are present in parser
  #print(soup) #tree ke form not printed but inside its tree
  #niche wale mein find_all finds div cells jinki property attrs mein given hai
  data=[] #took a list for storing data at end
  for d in soup.find_all('div',attrs={'class':"zg-grid-general-faceout"}):
  #we have to convert the class into dictionary key pair value to provide
  #now ek ek krke har ek div we will pick
  #once we have located the div, we need writer name, name of book,price,link of book etc.
  #suppose lets go to book name and inspect krke check book name ka div ...oski class property uthao
  #mostly hum sab find_all se sab data nikal lete hai
    book_name=d.find('div',{'class':"_cDEzb_p13n-sc-css-line-clamp-1_1Fn1y"}) #basically d cell mein jaha par yeh class wlai property first satisfied voh book text mil jayega..sab books ke names mil jayenge
  #author name is exactly same as book name
  #toh book names aa jayenge ...ya  tph find next krlo
  #ya phir oska parent class dhondlo...voh div same header ko highlighting
  #aab parent mein div mein no content and oske niche jitne bhi baaki content voh display ho jayega
    author_name = d.find('div',{'class':"a-row a-size-small"})
  #now stars ki dekho
    ratings = d.find('span',{'class':"a-icon-alt"}) #span tag mein likha hai 5 out of 5 stars isliye voh used
  #now price
    price = d.find('span',{'class':"p13n-sc-price"}) #jo original property dalenge then price wala nhi aayega ...even though its not same with any other class
    #sometimes websites ne khud se kuch kra hota hai...to block price checkers to block price checkers
    #in this try by removing things after underscore and it will work..remove underscore also
    #yeh class names are changed for web security
  #now votes
    votes = d.find('span',{'class':"a-size-small"})
  #now book type
    book_type = d.find('span',{'class':"a-size-small a-color-secondary a-text-normal"})
  #now link
    link = d.find('a',{'class':"a-link-normal"})
    #print(book_name.text) #yeh tab use kiya when did book name
    #print(author_name.text)
    #print(ratings.text) #attribute error might come ..i.e because kisi book ki rating missing hogi..we will handle that also ya phir sometimes request jake retrieve nhi ya time lga deta hai...tsab bhi run time error dega
    #print(votes.text) #pre processing ke vakat aise error aayenge..jaise beech mein names aa rhe onke same span tag mein property hogi and voh phele retrieve ho gya
    #print(book_type.text)
    #print(link.text) isse nothing u will get since link href mein hota hai not in anchor tag...not use get to get the property of tag
    #print(link.get("href")) #but isse links aa jayenge but links open nhi honge since base url nhi hia...hume base url append krna pdega
    #print('https://www.amazon.in'+ link.get('href'))

    #now we have got everything from d...now suppose we have some error if something is missing
    #ya sometimes beautiful soup ka error aa jayega because connection breaks toh run time error
    #toh aab isko dekho
    name = book_name.text if book_name else 'nan' #toh now we are storing the text varna not a number kr denge
    author = author_name.text if author_name else 'nan'
    rating = ratings.text if ratings else 'nan'
    vote = votes.text if votes else 'nan'
    type_book = book_type.text if book_type else 'nan'
    amount = price.text if price else 'nan'
    book_link = 'https://www.amazon.in'+ link.get('href') if link else 'nan'

    #now we will append everything into dictionary....now will make database of the data retrieved
    data.append({
        'name':name,
        'author':author,
        'rating':rating,
        'vote':vote,
        'type_book':type_book,
        'amount':amount,
        'book_link':book_link
    })

  return data
```

```python
#get_data(1) #when were printing upar then isko use krke dekh rhe the what was getting printed and now we got a lists of dictionaries
#now we can convert this into database pandas ka
import pandas as pd
pd.DataFrame(get_data(1))
```

```python
#now suppose we need data from multiple pages
#suppose 3 pages se data ekatha krna hai
no_of_pages = 3
all_data=[]
for pageno in range(1,no_of_pages+1): #page number start from 1 so..
  all_data.extend(get_data(pageno))
#converting all data into dataframe
#since bas 2 pages the toh pura data hi aa gya
df=pd.DataFrame(all_data)
#now isko we can store in it anywhere in the system in form of excel sheet
df.to_csv(r'') #in '' put path of your computer jaha dalna hai
```
