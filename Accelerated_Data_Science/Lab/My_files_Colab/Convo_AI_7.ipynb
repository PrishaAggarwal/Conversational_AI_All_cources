```python
def get_data(API_key,year,genre_id,sort_value,pageno): #we are making a general fucntion for it
  #yeh sab parameters are in discover toh we will use that
  url=base_url+'/discover/movie?api_key='+API_key+'&primary_release_year.gte='+str(year)+'&with_genres='+ str(genre_id)+'&sort_by='+sort_value+'&page='+str(pageno)
  #complete url must be in string format
  #every filter is concatenated with '&' sign once after ? i.e query u have given api key for authentication
  #'&' ke baad we have written the parameter name given in documentation
  #.gte is greater than equal to
  #u can direcly give comedy here directly but hum ek general function bana rhe hai jisse whatever we need we will get
  response=requests.get(url)


  #since we are gonna retrieve data again and again so we have to should either do something about error
  #aab in documentation u can see status code is 200
  #toh 1st method mein do agar status code is 200 then retrieve else dont retrieve

  #2nd method u can use sleep parameter from selenium
  #i.e after every request u can make retrieving of data for some time

  #u can see reponse body that what format mein and what keys/values return hongi

  if response.status_code == 200:
    data=response.json() #response ek json file hai jsiemin results mein sabhi mvoie ki  info hai
    return data['results'] #json file ko convert krke we need that results wali key from data toh returning that
  else:
      return 'error'
```

```python
get_data(API_key,2000,35,'popularity.desc',1) #popularity.desc documentation se dekhkr likha hai
#page no 1 se we have taking suppose
#so now we can use it for sentiment analysis dataset
```

```python
#now suppose we want 300 files ka data and osko we will store in dataframe format
import pandas as pd
df=pd.DataFrame(columns=['movie_id','movie_name']) #hume bas 2 cheeze store karvani hai as mentioned in ques
#we have to call function data tab tak jab tak length of rows doesnt become 300
#initialize the pageno from 1
pageno=1
while(len(df)<300):
  query_results=get_data(API_key,2000,35,'popularity.desc',pageno)
  #aab dictionary as well as error mil sakta hai and we have to deal with error as well
  if query_results == 'error':
    continue
  #i will get list of dictionaries
  #hum osmein se ek ek result uthayeneg and df mein new row add kr denge
  #os row mein we will store id and saath mein chahe store original title or title
  for movie in query_results:
    df.loc[len(df)]=[movie['id'],movie['title']]
    #whenever empty dataframe mein u have to store toh u take len(df) as index
  pageno+=1
```

```python
#we have method for similar movies
#chahe page no mat lo since ek page se 5 similar mvoie nikal hi jayengi..and by default pageno is 1
#api key and get movie id jiske correponding 5 movies nikalni hai
#movie ke  under similar ke under u will find the method url
def get_similar_movies(API_key,movie_id):
  url=base_url+'/movie/'+str(movie_id)+'/similar?api_key='+API_key
  response = requests.get(url)
  print(response.status_code)
  if response.status_code == 200:
    data=response.json()
    similar_movie_ids = []
    #data mein results jo return krega omsein se we only want id so take a dictionary for it
    for movie in data['results']:
      similar_movie_ids.append(movie['id']) # we will append movie key ki id ki value
    return similar_movie_ids
  else:
    return 'error'
```
