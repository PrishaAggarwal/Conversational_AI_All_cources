```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_iris
dataset=load_iris()

X=dataset.data
Y=dataset.target
```

```python
df=pd.read_csv(r"C:\Users\jasme\OneDrive\Desktop\LDA_1.csv")
X=np.array(df.iloc[:,:-1])
Y=np.array(df.iloc[:,-1])
```

```python
def between_scatter(X,Y):
    n,k=X.shape
    global_mean=X.mean(axis=0)
    Sb=np.zeros((k,k),dtype=float)
    classes,count=np.unique(Y,return_counts=True)
    for cls in classes:
        mask = (Y==cls)
        X_subset=X[mask]
        class_mean=X_subset.mean(axis=0)
        n_class=X_subset.shape[0]
        diff=(class_mean-global_mean).reshape(1,k)
        #diff = diff.reshape(k, 1)
        #Sb += n_class * (diff * diff.T)
        Sb+=n_class * diff.T * diff
    return Sb
    
```

```python
def within_class(X,Y):
    n,k=X.shape
    classes,count=np.unique(Y,return_counts=True)
    Sw=np.zeros((k,k),dtype=float)
    for cls in classes:
        mask=(Y==cls)
        X_subset=X[mask]
        class_mean=(X_subset.mean(axis=0)).reshape(1,k)
        diff=X_subset-class_mean
        #print(diff.shape)
        Sw+=diff.T@diff
    return Sw
    
```

```python
Sb=between_scatter(X,Y)
Sw=within_class(X,Y)
cov=np.linalg.inv(Sw)@Sb
values,vectors=np.linalg.eig(cov)
```
