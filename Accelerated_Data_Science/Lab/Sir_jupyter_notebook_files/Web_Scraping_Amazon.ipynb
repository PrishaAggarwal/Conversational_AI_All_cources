```python
import requests
from bs4 import BeautifulSoup
```

```python
def get_data(pageno):
    url='https://www.amazon.in/gp/bestsellers/books/ref=zg_bs_pg_'+str(pageno)+'_books?ie=UTF8&pg='+str(pageno)
    r=requests.get(url)
    soup=BeautifulSoup(r.content,'html.parser')
    data=[]
    for d in soup.find_all('div',attrs={'class':"p13n-sc-uncoverable-faceout"}):
        book_name=d.find('div',attrs={'class':"_cDEzb_p13n-sc-css-line-clamp-1_1Fn1y"})
        author_name=d.find('div',attrs={'class':"a-row a-size-small"})
        ratings=d.find('span',attrs={'class':"a-icon-alt"})
        no_of_votes=d.find('span',attrs={'class':"a-size-small"})
        book_type=d.find('span',attrs={'class':"a-size-small a-color-secondary a-text-normal"})
        price=d.find('span',attrs={"class": "p13n-sc-price"})
        book_link=d.find('a',{'class':"a-link-normal aok-block"})
        name=book_name.text if book_name else 'NaN'
        author=author_name.text if author_name else 'NaN'
        rating=ratings.text if ratings else 'NaN'
        votes=no_of_votes.text if no_of_votes else 'NaN'
        Type=book_type.text if book_type else 'NaN'
        Amount=price.text if price else 'NaN'
        link='https://www.amazon.in'+book_link.get('href') if book_link else 'NaN'
        data.append({'Name':name,'Author':author,'Stars':rating,'Votes':votes,'Type':Type,'Price':Amount,'Link':link})
    return data
        
    
    
```
